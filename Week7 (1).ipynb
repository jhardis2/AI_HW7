### This project explored which clinical and lifestyle features most strongly differentiate chronic kidney disease (CKD) patients from non-CKD patients using a combination of exploratory data analysis (EDA), visual storytelling, and predictive modeling.

### 1. Serum Creatinine Trends Over Time
A highlighted line chart was used to visualize a simulated time series of serum creatinine (SCR) levels. A sharp spike in recent values was emphasized using color and annotation, mirroring the techniques in *Storytelling With Data* (Chapter 4). This chart demonstrates how storytelling techniques can guide the viewers attention to a potential clinical event or turning point.

**Key Insight:** Rising serum creatinine over time is a recognizable early signal of CKD progression.

---

### 2. Feature Differences Between CKD and Non-CKD Groups
Boxplots comparing CKD and non-CKD patients revealed:

- **Serum Creatinine:** Significantly higher in CKD patients, with a wider range of values.
- **Hemoglobin:** Lower in CKD patients, reflecting anemia as a common comorbidity.

These visual comparisons clearly show which features are strong candidates for prediction and clinical screening.

---

### 3. Feature Importance (Model-Based Insight)
A simulated machine learning feature importance chart showed:

- **Top Predictors:** `serum_creatinine`, `hemoglobin`, and `albumin`
- **Secondary Predictors:** `glucose`, `hypertension`, and `diabetes`

This reinforces both clinical knowledge and supports model-driven decision-making by focusing on the most informative features.

---

### 4. Correlation Between Clinical Features
A heatmap of feature correlations revealed:

- A **strong negative correlation** between serum creatinine and both hemoglobin and albumin.
- Positive correlation between hemoglobin and albumin, suggesting a shared physiological link.

Understanding these correlations helps reduce multicollinearity in models and improves feature selection.

---

###  Summary of Key Findings:
- CKD patients show **higher SCR**, **lower hemoglobin**, and **increased prevalence of comorbidities** like hypertension and diabetes.
- **Data storytelling methods** (annotation, color, layout) made complex clinical trends more accessible.
- These findings directly support the goal of building models to **predict CKD using accessible, interpretable features**.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Simulated CKD dataset for demonstration purposes
np.random.seed(42)
n_ckd = 100
n_non_ckd = 100

data = {
    'classification': ['ckd'] * n_ckd + ['notckd'] * n_non_ckd,
    'serum_creatinine': np.concatenate([np.random.normal(3.0, 0.8, n_ckd), np.random.normal(1.0, 0.3, n_non_ckd)]),
    'hemoglobin': np.concatenate([np.random.normal(10, 1.5, n_ckd), np.random.normal(13.5, 1.0, n_non_ckd)]),
    'albumin': np.concatenate([np.random.normal(3.0, 0.5, n_ckd), np.random.normal(4.5, 0.3, n_non_ckd)]),
    'glucose': np.concatenate([np.random.normal(150, 40, n_ckd), np.random.normal(110, 25, n_non_ckd)]),
    'hypertension': [1]*70 + [0]*30 + [1]*40 + [0]*60,
    'diabetes': [1]*60 + [0]*40 + [1]*30 + [0]*70
}

ckd_df = pd.DataFrame(data)

# Simulate a serum creatinine time series for one patient
scr_time = pd.date_range(start='2023-01-01', periods=60)
scr_values = np.concatenate([np.random.normal(1.0, 0.2, 45), np.random.normal(3.5, 0.3, 15)])
scr_df = pd.DataFrame({'date': scr_time, 'serum_creatinine': scr_values})

# Create visualizations
figures = []

# 1. Highlighted SCR Line Chart
plt.figure(figsize=(10, 5))
plt.plot(scr_df['date'][:45], scr_df['serum_creatinine'][:45], color='lightgray', label='Earlier')
plt.plot(scr_df['date'][45:], scr_df['serum_creatinine'][45:], color='blue', linewidth=2.5, label='Recent Spike')
plt.annotate('Spike in SCR',
             xy=(scr_df['date'][50], scr_df['serum_creatinine'][50]),
             xytext=(scr_df['date'][50], scr_df['serum_creatinine'][50] + 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10, bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.3))
plt.title('Serum Creatinine Over Time')
plt.xlabel('Date')
plt.ylabel('Serum Creatinine (mg/dL)')
plt.xticks(rotation=45)
plt.tight_layout()
figures.append(plt.gcf())

# 2. Boxplot: Serum Creatinine by CKD Status
plt.figure(figsize=(8, 5))
sns.boxplot(data=ckd_df, x='classification', y='serum_creatinine', palette='Set2')
plt.title('Serum Creatinine by CKD Status')
plt.xlabel('CKD Classification')
plt.ylabel('Serum Creatinine (mg/dL)')
plt.tight_layout()
figures.append(plt.gcf())

# 3. Boxplot: Hemoglobin by CKD Status
plt.figure(figsize=(8, 5))
sns.boxplot(data=ckd_df, x='classification', y='hemoglobin', palette='Set2')
plt.title('Hemoglobin Levels by CKD Status')
plt.xlabel('CKD Classification')
plt.ylabel('Hemoglobin (g/dL)')
plt.tight_layout()
figures.append(plt.gcf())

# 4. Feature Importance Bar Chart (Simulated Importance Scores)
feature_importance = pd.Series({
    'serum_creatinine': 0.35,
    'hemoglobin': 0.25,
    'albumin': 0.20,
    'glucose': 0.10,
    'hypertension': 0.07,
    'diabetes': 0.03
})

plt.figure(figsize=(8, 5))
feature_importance.sort_values().plot(kind='barh', color='steelblue')
plt.title('Top Predictive Features for CKD (Simulated)')
plt.xlabel('Importance Score')
plt.tight_layout()
figures.append(plt.gcf())

# 5. Heatmap of Feature Correlations
plt.figure(figsize=(8, 6))
corr = ckd_df[['serum_creatinine', 'hemoglobin', 'albumin', 'glucose']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Clinical Features')
plt.tight_layout()
figures.append(plt.gcf())
plt.savefig("Heatmap_of_Feature_Correlations_week7.png", dpi=300, bbox_inches='tight')

ckd_df.head()
```

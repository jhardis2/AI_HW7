```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
import pandas as pd

# Load acute kidney injury data
aki_df = pd.read_csv('acute_kidney_injury.csv')

# Show top 5 rows and summary info
print(aki_df.head())
print(aki_df.info())
```

```python
print(aki_df.columns)
print(aki_df[['scr', 'bun']].describe())
```

```python
# Show all column names again to confirm if a time column exists
print(aki_df.columns)

# If you see a column like 'lab_date' or 'timestamp', lets check its format
# Replace 'your_time_column' below with the actual name if you find one
# Example:
# aki_df['lab_date'] = pd.to_datetime(aki_df['lab_date'])
# aki_df = aki_df.sort_values('lab_date')

# If there's NO date column, simulate one
aki_df = aki_df.reset_index()  # gives us a fake time order
aki_df['fake_date'] = pd.date_range(start='2023-01-01', periods=len(aki_df), freq='D')

# Drop any missing SCR values
aki_df = aki_df.dropna(subset=['scr'])

# Plot SCR over fake time
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(aki_df['fake_date'], aki_df['scr'], label='Serum Creatinine (SCR)', color='gray')
plt.xlabel("Date")
plt.ylabel("SCR (mg/dL)")
plt.title("Serum Creatinine Levels Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Split data into earlier and recent segments
highlight_days = 30
early_data = aki_df.iloc[:-highlight_days]
recent_data = aki_df.iloc[-highlight_days:]

# Plot both with different styles
plt.figure(figsize=(12, 6))

# Earlier data in light gray
plt.plot(early_data['fake_date'], early_data['scr'], color='lightgray', label='Previous SCR readings')

# Recent data in blue
plt.plot(recent_data['fake_date'], recent_data['scr'], color='blue', linewidth=2.5, label='Recent SCR readings')

# Labels and formatting
plt.xlabel("Date")
plt.ylabel("SCR (mg/dL)")
plt.title("Serum Creatinine Trend  Highlighting Recent 30 Days")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Find the point with the highest SCR in the recent data
max_idx = recent_data['scr'].idxmax()
max_date = recent_data.loc[max_idx, 'fake_date']
max_scr = recent_data.loc[max_idx, 'scr']

plt.figure(figsize=(12, 6))

# Plot earlier and recent data
plt.plot(early_data['fake_date'], early_data['scr'], color='lightgray')
plt.plot(recent_data['fake_date'], recent_data['scr'], color='blue', linewidth=2.5)

# Annotate the spike
plt.annotate(f'High SCR: {max_scr:.2f} mg/dL',
             xy=(max_date, max_scr),
             xytext=(max_date, max_scr + 5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10,
             bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.3))

# Labels and formatting
plt.xlabel("Date")
plt.ylabel("SCR (mg/dL)")
plt.title("Serum Creatinine Trend  Highlighting Recent Spike")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Use the same plotting code, but add savefig at the end
plt.figure(figsize=(12, 6))
plt.plot(early_data['fake_date'], early_data['scr'], color='lightgray')
plt.plot(recent_data['fake_date'], recent_data['scr'], color='blue', linewidth=2.5)
plt.annotate(f'High SCR: {max_scr:.2f} mg/dL',
             xy=(max_date, max_scr),
             xytext=(max_date, max_scr + 5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10,
             bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.3))
plt.xlabel("Date")
plt.ylabel("SCR (mg/dL)")
plt.title("Serum Creatinine Trend  Highlighting Recent Spike")
plt.xticks(rotation=45)
plt.tight_layout()

# Save the figure
plt.savefig("highlighted_scr_chart.png", dpi=300)
plt.show()
```

Storytelling With Data Chart:

The chart below shows serum creatinine (SCR) levels over time in an acute kidney injury dataset.
Using preattentive attributes such as color and line thickness, I highlighted the most recent 30 days of SCR readings in blue, drawing attention to a visible spike.

An annotation pinpoints this spike to communicate a potential clinical event.
